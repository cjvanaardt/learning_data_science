### Scope
- Mutable objects passed into a function can be changed within the function
- Using global when declaring a variable inside a function gives access to the variable in a lower scope

```python
data = [1, 2, 3]

num = 4

def func1(data, val):
    data.append(val)
def func2(data, val):
    num = 3
def func3(data, val):
    global num
    num = 3

func1(data, val)
print(data)
func2(data, val)
print(num)
func3(data, val)
print(num)
```

#### Dates and Times
- The datetime module provides datetime, date and time types
- date and time types can be extracted from datetime types
- datetime types have a specific string formatting method
- the difference between two datetime types results in a timedelta type
- timedelta types can be added to datetime types

```python
from datetime import datetime, date, time

# datetime type
dt = datetime(2011, 10, 29, 20, 30, 21)
(dt.year, dt.month, dt.day, dt.hour, dt.minute, dt.second)
```

```python
# date type
z = date(2011, 10, 29)
(z.year, z.month, z.day)
```

```python
# time type
y = time(20, 30, 21)
(y.hour, y.minute, y.second)
```

```python
# extracting date and time types
dt.date()
dt.time()
```

```python
# datetime string formatting
z.strftime('%Y/%m/%d')
y.strftime('%H:%M:%S')
dt.strftime('%Y/%m/%d %H:%M:%S')
```

```python
# difference in datetime objects
dt2 = datetime(2012, 11, 30, 21, 31, 22)
diff = dt - dt2

dt2 + diff
```

### Control Flow
- Conditions are expressions that evaluate down to a boolean value
- The control statements if, elif and else are used to control the flow of python programs using conditions
- The block of code following is called a clause
- for loops iterate over a collection or a iterator 
- while loops continue looping until their condition becomes false
- break statements exit the current loop
- pass continues execution as if nothing happened
- continue sends a loop back to the next loop
- sys.exit() will end the program
- Ctrl-C will exit an infinite loop

```python
# if, elif, else
if 0 > 3:
    print("That's not right!")
elif 0 > 2:
    print("That's definitely not right!")
elif 0 > 1:
    print("This must be a joke!")
else:
    print("So 0 <= 1")
```

```python
for i in range(5):
    print(i)
```

```python
i = 0
while i < 5:
    print(i)
    i += 1
```
